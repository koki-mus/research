```python
import py_compile
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import cv2
import os
from dotenv import load_dotenv
import glob 
from mymodule import myfunc as mf
load_dotenv()
imgout = os.environ["imgout"]
root_dir = os.environ["root_dir"]
snaps_dir = os.environ["snaps_dir"]
```

```python
path_list = glob.glob(snaps_dir+"enstrophy/*/*")
num_of_data = len(path_list) # リコネクションがない画像の枚数

def compress(array, LEVEL=10):
    return mf.convolute(array,mf.ave_carnel(LEVEL), stride = LEVEL)
temp_data = compress(mf.load(path_list[0],z=3))
IMGSHAPE = temp_data.shape

N_col = IMGSHAPE[0]*IMGSHAPE[1]*1 # 行列の列数
X_train = np.zeros((num_of_data, N_col)) # 学習データ格納のためゼロ行列生成
y_train = np.zeros((num_of_data)) # 学習データに対するラベルを格納するためのゼロ行列生成

# リコネクションがない画像を行列に読み込む
i_count = 0

def load_regularize(item):
    type = item[:-4]
    print(item)
    if type == ".npy":
        im = np.load(item)
    elif type == ".npz":
        print("npz doesnot supported")
        return
    else:
        im = mf.load(item, z=3)
    img_resize = compress(im)
    return ((img_resize - min(img_resize.flat)) / max(img_resize.flat)).flat # 正規化

for item in path_list:
    X_train[i_count,:] = load_regularize(item)
    y_train[i_count] = 0 # リコネクションがないことを表すラベル
    i_count += 1
```

```python
#PCA
from sklearn.decomposition import PCA
N_dim =  100 # 49152(=128×128×3)の列を100列に落とし込む
pca = PCA(n_components=N_dim, random_state=0)
X_train_pca = pca.fit_transform(X_train)
print('累積寄与率: {0}'.format(sum(pca.explained_variance_ratio_)))
```
